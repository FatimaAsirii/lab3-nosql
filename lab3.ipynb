```python
from pymongo import MongoClient
from pymongo.server_api import ServerApi
import certifi

MONGO_URI = "mongodb+srv://<username>:<password>@cluster0.ivxxhsi.mongodb.net/?appName=Cluster0"

client = MongoClient(
    MONGO_URI,
    server_api=ServerApi("1"),
    tls=True,
    tlsCAFile=certifi.where(),
)

client.admin.command("ping")
print("Atlas connected successfully!")
```

```python
db = client["ds_lab_mongo"]
customers = db["customers"]

customers.delete_many({})  # clean start
print("Documents in customers:", customers.count_documents({}))
```

```python
result = customers.insert_one({
    "name": "Alice",
    "email": "alice@email.com",
    "country": "USA"
})

print("Inserted _id:", result.inserted_id)
print("Documents in customers:", customers.count_documents({}))
```

```python
customers.insert_one({
    "name": "David",
    "email": "david@email.com",
    "country": "UK"
})

print("Documents in customers:", customers.count_documents({}))
```

```python
uk_customers = list(customers.find({"country": "UK"}, {"_id": 0}))
print(uk_customers)
```

```python
update_result = customers.update_one(
    {"name": "Alice"},
    {"$set": {"email": "alice_updated@email.com"}}
)

print("Matched:", update_result.matched_count)
print("Modified:", update_result.modified_count)
print(customers.find_one({"name": "Alice"}, {"_id": 0}))
```

```python
customers.insert_one({
    "name": "Charlie",
    "email": "charlie@email.com",
    "country": "UK"
})

delete_result = customers.delete_one({"name": "Charlie"})
print("Deleted:", delete_result.deleted_count)
print("Documents in customers:", customers.count_documents({}))
```

```python
customers.insert_one({
    "name": "Bob",
    "email": "bob@email.com",
    "country": "UK",
    "address": {
        "city": "London",
        "zip": "SW1A"
    }
})

print("Inserted document with nested address successfully.")
```

```python
london_customers = list(customers.find({"address.city": "London"}, {"_id": 0}))
print(london_customers)
```
